```python
with open('./data/recipes.json') as f: #        "f"
    recipes = json.load(f)

#pprint(recipes)
#pprint(recipes[0]['ingredients'])
for i in recipes:
    if i['id'] == 13121:
        pprint(i)
```

```python
cuisine = set()
for i in recipes:
    cuisine.add(i['cuisine'])
    #pprint(i['cuisine'])
print(len(cuisine))
```

```python
count = 0
for i in cuisine:
    for j in recipes:
        if i == j['cuisine']:
            count += 1
    print(f'{ i } - { count }')
    count = 0
```

```python
df = pd.DataFrame(recipes) #   DataFrame   recipes
display(df.head())
```

```python
all_ingredients=set() #        
for recipe in recipes: #       
    for ingredient in recipe['ingredients']: #    ,     
        all_ingredients.add(ingredient) #     
display(len(all_ingredients))
```

```python
def contains(ingredient_list): #      
    if ingredient_name in ingredient_list: #      ,
        return 1 #   1
    else: #      ,
        return 0 #   0
for ingredient_name in all_ingredients:    
    df[ingredient_name] = df['ingredients'].apply(contains)
```
