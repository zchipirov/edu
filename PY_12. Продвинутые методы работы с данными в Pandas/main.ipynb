```python
import pandas as pd

melb_data = pd.read_csv('./data/melb_data_fe.csv', sep=',')
```

```python
melb_df = melb_data.copy()
```

```python
melb_df.head()
```

```python
melb_df.info()
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'], errors='coerce')
```

```python
melb_df['Date'].dt.quarter.value_counts()
```

```python
exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']

for col in melb_df.columns:
    if melb_df[col].nunique() < 150 and col not in exclude:
        #print(col, melb_df[col].nunique())
        melb_df[col] = melb_df[col].astype('category')
```

```python
melb_df.info()
```

```python
melb_df.sort_values(by='Price').head(10)
```

```python
melb_df.sort_values(by='Date', ascending=False)
```

```python
melb_df.sort_values(by=['Distance', 'Price']).loc[::10, ['Distance', 'Price']]
```

```python
mask1 = melb_df['AreaRatio'] < -0.8
mask2 = melb_df['Type'] == 'townhouse'
mask3 = melb_df['SellerG'] == 'McGrath'
melb_df[mask1 & mask2 & mask3].sort_values(
    by=['Date', 'AreaRatio'],
    ascending=[True, False],
    ignore_index=True
).loc[:, ['Date', 'AreaRatio']]
```

```python
melb_df['AreaRatio']
```

```python
melb_df.sort_values(by='AreaRatio', ascending=False, ignore_index=True).loc[::1558, ['Distance', 'BuildingArea', 'AreaRatio']]
```

```python
melb_df[(melb_df['Type'] == 'townhouse') & (melb_df['Rooms'] > 2)].sort_values(by=['Rooms', 'MeanRoomsSquare'], ascending=[True, False], ignore_index=True).loc[18]
```

```python
melb_df.groupby('Rooms')['Price'].mean().sort_values(ascending=False)
```

```python
melb_df.groupby('Regionname')['Lattitude'].std()
```

```python
melb_df[(melb_df['Date'] > '2017-05-01') & (melb_df['Date'] <= '2017-09-01')].groupby('SellerG')['Price'].sum().sort_values(ascending=True)
```

```python
melb_df.groupby(['Rooms', 'Type'])['Price'].mean().unstack()
```

```python
melb_df.pivot_table(
    values='BuildingArea',
    index='Rooms',
    columns='Type',
    aggfunc='median',
    fill_value=0
)
```

```python
unit = melb_df.pivot_table(
    values='Price',
    index='SellerG',
    columns='Type',
    aggfunc='median',
    fill_value=0
)
display(unit['unit'].sort_values(ascending=False))
```
